Forcing Extremes
* Goal: Visually observe underfitting and overfitting.
* Task: Generate a non-linear regression dataset using make_regression or mathematical functions (like a noisy sine wave).
* Deliverable: Train three DecisionTreeRegressor models: one with max_depth=1 (high bias), one with max_depth=3 (balanced), and one with max_depth=None (high variance). Plot the true data alongside the prediction lines of all three models on the same graph using Matplotlib or Seaborn.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# 1. Generate noisy, curved data (similar to your slides)
np.random.seed(42)
X = np.sort(10 * np.random.rand(30, 1), axis=0)
y = (-0.15 * (X - 3)**2 + 5).ravel() + np.random.normal(0, 0.4, X.shape[0])

# Line points for drawing predictions smoothly
X_grid = np.linspace(0, 10, 500)[:, np.newaxis]

# 2. Train the three models
tree_underfit = DecisionTreeRegressor(max_depth=1).fit(X, y)
tree_balanced = DecisionTreeRegressor(max_depth=3).fit(X, y)
tree_overfit  = DecisionTreeRegressor(max_depth=None).fit(X, y)

# 3. Plot everything together
plt.scatter(X, y, color="blue", label="Data")
plt.plot(X_grid, tree_underfit.predict(X_grid), color="orange", label="Underfit (Depth 1)")
plt.plot(X_grid, tree_balanced.predict(X_grid), color="green", label="Balanced (Depth 3)")
plt.plot(X_grid, tree_overfit.predict(X_grid), color="red", label="Overfit (Depth Max)")

plt.legend()
plt.show()
```
